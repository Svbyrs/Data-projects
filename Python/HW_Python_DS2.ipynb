```python
# 1 TASK (Исследование нормальных распределений)

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate two normal distributions
np.random.seed(42)
standard_normal = np.random.normal(loc=0, scale=1, size=1000)
shifted_normal = np.random.normal(loc=8, scale=2, size=1000)
```

```python
# Step 2: Plot normalized histograms for both distributions
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(standard_normal, bins=30, density=True, alpha=0.6, color='blue', label='Standard Normal (μ=0, σ=1)')
plt.hist(shifted_normal, bins=30, density=True, alpha=0.6, color='orange', label='Normal (μ=8, σ=2)')
plt.title('Original Distributions')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Step 3: Standardize the second distribution
standardized_shifted = (shifted_normal - np.mean(shifted_normal)) / np.std(shifted_normal)
```

```python
# Step 4: Plot histogram of standardized data vs standard normal
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 2)
plt.hist(standard_normal, bins=30, density=True, alpha=0.6, color='blue', label='Standard Normal (μ=0, σ=1)')
plt.hist(standardized_shifted, bins=30, density=True, alpha=0.6, color='green', label='Standardized (μ≈0, σ≈1)')
plt.title('Standardized Comparison')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 2 TASK (Изучение распределения роста учащихся средней школы и его стандартизация для анализа.)

# Step 1: Generate original height data
np.random.seed(42)
original_heights = np.random.normal(loc=170, scale=10, size=1000)
```

```python
# Step 2: Standardize the data
standardized_heights = (original_heights - np.mean(original_heights)) / np.std(original_heights)
```

```python
# Step 3: Plot histograms
plt.figure(figsize=(12, 5))

# Histogram of original heights
plt.subplot(1, 2, 1)
plt.hist(original_heights, bins=30, density=True, color='skyblue', edgecolor='black')
plt.title('Original Heights Distribution')
plt.xlabel('Height (cm)')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
# Histogram of standardized heights
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 2)
plt.hist(standardized_heights, bins=30, density=True, color='lightgreen', edgecolor='black')
plt.title('Standardized Heights Distribution')
plt.xlabel('Standardized Height')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
# 3 TASK (Изменение климата приводит к тому, что погодные условия становятся более переменчивыми, что включает изменения в среднесуточных температурах. В этом задании вы будете работать с моделью данных среднесуточных температур для определенного месяца, например, июля. Исходя из исторических данных, предположим, что средняя температура в июле составляет 22°C со стандартным отклонением 3°C.)

# Step 1: Generate 1000 daily average temperatures for July
np.random.seed(42)
temperatures = np.random.normal(loc=22, scale=3, size=1000)
```

```python

# Step 2: Calculate percentage of anomalously high (>27°C) and low (<15°C) temperatures
high_anomalies = temperatures > 27
low_anomalies = temperatures < 15

percent_high = np.sum(high_anomalies) / len(temperatures) * 100
percent_low = np.sum(low_anomalies) / len(temperatures) * 100
```

```python
# Step 3: Plot histogram with highlighted anomaly regions
bins = np.linspace(min(temperatures), max(temperatures), 30)
plt.hist(temperatures[~(high_anomalies | low_anomalies)], bins=bins, color='skyblue', label='Normal', alpha=0.7)
plt.hist(temperatures[high_anomalies], bins=bins, color='red', label='> 27°C (High)', alpha=0.7)
plt.hist(temperatures[low_anomalies], bins=bins, color='blue', label='< 15°C (Low)', alpha=0.7)

plt.xlabel('Daily Average Temperature (°C)')
plt.ylabel('Frequency')
plt.title('Distribution of July Daily Average Temperatures\nwith Anomalies Highlighted')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Print the calculated percentages
print(f"Percentage of days with temperature > 27°C: {percent_high:.2f}%")
print(f"Percentage of days with temperature < 15°C: {percent_low:.2f}%")
```
