```python
#1 STEP
import pandas as pd
```

```python
# Загрузка набора данных
data = pd.read_csv('CSE_student_performances.csv')
```

```python
# Просмотр первых нескольких строк набора данных
print("Первые несколько строк данных:")
data.head()
```

```python
# Получение информации о данных
print("\nИнформация о данных:")
data.info()
```

```python
# Описательная статистика
print("\nОписательная статистика:")
data.describe()
```

```python
#2 STEP
# Удаление колонки 'LikePresentations'
data = data.drop(columns=['LikePresentation'])
```

```python
# Проверка, что колонка удалена
print("\nПервые несколько строк данных после удаления колонки 'LikePresentation':")
data.head()
```

```python
#3 STEP
import re
# Функция для вставки пробелов перед заглавными буквами
def split_camel_case(name):
    return re.sub(r'(?<=[a-z])(?=[A-Z])', ' ', name)

# Применим к каждому названию столбца
data.columns = [split_camel_case(col) for col in data.columns]
```

```python
# Проверка новых названий колонок
print("\nНовые названия колонок:")
data.columns
```

```python
# Проверка первых нескольких строк данных после изменений
print("\nПервые несколько строк данных после изменений:")
data.head()
```

```python
#4 STEP
# Удаление дубликатов
data.drop_duplicates(inplace=True)
```

```python
# Удаление строк с пропущенными значениями
data.dropna(inplace=True)
```

```python
# Проверка результата
print("Размер данных после очистки:")
data.shape
```

```python
#5 STEP
# среднее количество сна студентов
data['Sleep Per Day Hours'].mean()
```
