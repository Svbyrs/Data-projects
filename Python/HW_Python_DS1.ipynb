```python
# 1 TASK

import numpy as np

# Step 1: Generate a list of 1000 random integers between 1 and 6
dice_rolls = np.random.randint(1, 7, size=1000)

# Display the first 20 rolls as a sample
print("Sample of 20 dice rolls:", dice_rolls[:20])
```

```python
import matplotlib.pyplot as plt

# Step 2 & 3: Plot histogram with adjusted parameters
plt.hist(dice_rolls, bins=np.arange(1, 8) - 0.5, rwidth=0.8, density=True, edgecolor='black')

# Step 4: Add axis labels and title
plt.xlabel('Dice Face')
plt.ylabel('Probability')
plt.title('Histogram of 1000 Dice Rolls (Uniform Distribution)')
plt.xticks(range(1, 7))
plt.grid(True)
plt.show()
```

```python
# 2 TASK ( Исследовать, как выбросы влияют на характеристики нормального распределения (среднее значение и стандартное отклонение), и визуализировать эффект выбросов на гистограмме распределения.)

# Step 1: Set seed for reproducibility
np.random.seed(42)
```

```python
# Step 2: Generate normal distribution data
original_data = np.random.normal(loc=50, scale=15, size=10000)                             
```

```python
# Step 3: Add outliers
outliers = np.full(10, 200)
modified_data = np.concatenate([original_data, outliers])
```

```python
# Step 4: Calculate new mean and standard deviation
new_mean = np.mean(modified_data)
new_std = np.std(modified_data)
```

```python
# Step 5: Visualize the modified distribution
plt.figure(figsize=(10, 6))
plt.hist(modified_data, bins=50, color='skyblue', edgecolor='black', density=True)
plt.axvline(new_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {new_mean:.2f}')
plt.axvline(new_mean + new_std, color='green', linestyle='dotted', linewidth=2, label=f'+1 Std: {new_mean + new_std:.2f}')
plt.axvline(new_mean - new_std, color='green', linestyle='dotted', linewidth=2, label=f'-1 Std: {new_mean - new_std:.2f}')
plt.title('Histogram of Modified Distribution with Outliers')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 3 TASK (Используйте KDE для анализа и сравнения распределения доходов жителей двух разных городов.)

import seaborn as sns

# Step 1: Generate income data for CityA and CityB
np.random.seed(42)
cityA_income = np.random.normal(loc=50000, scale=15000, size=1000)
cityB_income = np.random.normal(loc=55000, scale=20000, size=1000)
```

```python
# Step 2: Plot KDE for both cities
plt.figure(figsize=(10, 6))
sns.kdeplot(cityA_income, label='CityA', fill=True, color='blue')
sns.kdeplot(cityB_income, label='CityB', fill=True, color='orange')

# Step 3: Add vertical lines for mean income
plt.axvline(np.mean(cityA_income), color='blue', linestyle='--', linewidth=2, label='Mean CityA')
plt.axvline(np.mean(cityB_income), color='orange', linestyle='--', linewidth=2, label='Mean CityB')

# Step 4: Add title, labels, and legend
plt.title('Income Distribution Comparison: CityA vs CityB')
plt.xlabel('Income')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 4 TASK (Изучить распределение времени, необходимого для обслуживания клиентов в кафе, и проверить это распределение на "нормальность".)

import statsmodels.api as sm

# Step 1: Generate population data using exponential distribution
np.random.seed(42)
population = np.random.exponential(scale=10, size=1000)
```

```python
# Step 2: Draw a random sample of 100 values without replacement
sample = np.random.choice(population, size=100, replace=False)
```

```python
# Step 3: Visualize the sample distribution
plt.figure(figsize=(10, 5))
sns.kdeplot(sample, fill=True, label='KDE')
plt.hist(sample, bins=20, density=True, alpha=0.5, edgecolor='black', label='Histogram')
plt.title('Sample Distribution of Customer Service Times')
plt.xlabel('Service Time (minutes)')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 4: Q-Q plot to assess normality
sm.qqplot(sample, line='r')
plt.title('Q-Q Plot of Sample Service Times')
plt.grid(True)
plt.show()
```
