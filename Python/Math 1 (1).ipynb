```python
#1. Создайте функцию через def f(x), где на выходе вы получите x^4 + 3 * x^2 - 15
```

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return x**4 + 3 * x**2 - 15

# Пример использования функции
x_value = 2  
result = f(x_value)
print(f'f({x_value}) = {result}')

# 2. Создание точек для графика
x = np.linspace(-5, 5, 100)  # 100 точек от -5 до 5
y = f(x)

# 3. Построение графика
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', linewidth=2, label='f(x) = $x^4 + 3x^2 - 15$')
plt.title('График функции f(x) = $x^4 + 3x^2 - 15$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
# Создание списка x_points с числами от -5 до 5
x_points = list(range(-5, 6))

# Вывод списка
print(x_points)

# Строим график
plt.plot(x_points, x_points, marker='o')
plt.title("График точек от -5 до 5")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
#3. Используйте x_points как аргумент для созданной функции f(x), и выведите на экран результаты функции для каждого значения 
# x_points, также сохраните результаты в y_points
```

```python
# Создаем список x_points от -5 до 5
x_points = list(range(-5, 6))

# Определяем функцию f(x)
def f(x):
    return x**2 - 3*x + 2

# Вычисляем значения функции для каждого x и сохраняем в y_points
y_points = [f(x) for x in x_points]

# Выводим результаты
for x, y in zip(x_points, y_points):
    print(f"f({x}) = {y}")

# Строим график
plt.plot(x_points, y_points, marker='o')
plt.title("График функции f(x) = x^2 - 3x + 2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

```python
#4. Покажите на графике x_points & y_points
```

```python
# Создаем список x_points от -5 до 5
x_points = list(range(-5, 6))

# Определяем функцию f(x)
def f(x):
    return x**2 - 3*x + 2

# Вычисляем значения функции для каждого x и сохраняем в y_points
y_points = [f(x) for x in x_points]

# Строим график функции
plt.plot(x_points, y_points, label='f(x) = x^2 - 3x + 2')

# Добавляем маркеры для каждой точки
plt.scatter(x_points, y_points, color='red', zorder=5, label='x_points & y_points')

# Добавляем подписи и легенду
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции f(x) и точки (x, f(x))')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#5. Импортируйте функцию sin из библиотеки math (from math import sin), далее создайте список x_points, в котором лежат числа
# от -10 до 10. Аналогично создайте y_points используя функцию sin. Также отобразите на графике x_points & y_points
```

```python
from math import sin

# Создание списка x_points от -10 до 10
x_points = list(range(-10, 11))

# Вычисление y_points с использованием функции sin
y_points = [sin(x) for x in x_points]

# Построение графика
plt.plot(x_points, y_points, label='y = sin(x)', color='blue')
plt.scatter(x_points, y_points, color='red', label='Points')

# Добавление подписей и легенды
plt.title('График функции sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```
