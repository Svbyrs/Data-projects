**Упражнение 1:** Рассмотрим систему уравнений, приведенную ниже. Найдите такое значение коэффициента $a$, чтобы система имела ровно одно решение (то есть найти любое подходящее значение $a$). Для этого значения $a$ найдите решение системы уравнений и начертите линии, образованные вашими уравнениями, чтобы проверить свой ответ.

$$
\begin{eqnarray*}
ax_1 + 2x_2 & = & 1 \\
4x_1 + 8x_2 & = & 12 \\
\end{eqnarray*}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the coefficient matrix
# Equation 1: a*x1 + 2*x2 = 1
# Equation 2: 4*x1 + 8*x2 = 12

# We want to find a such that the determinant is non-zero
# Coefficient matrix:
# [[a, 2],
#  [4, 8]]

# Compute the determinant symbolically
from sympy import symbols, Eq, solve, Matrix

a = symbols('a')
matrix = Matrix([[a, 2], [4, 8]])
det = matrix.det()

# Solve for a where determinant is zero
a_singular = solve(Eq(det, 0), a)

# Choose a value of a that is not in the singular list
a_value = 2  # since a ≠ 1

# Now solve the system with a = 2
A = np.array([[a_value, 2], [4, 8]])
b = np.array([1, 12])

# Solve the system
solution = np.linalg.solve(A, b)

# Display results
print(f"Determinant expression: {det}")
print(f"Value(s) of a that make the determinant zero: {a_singular}")
print(f"Chosen a = {a_value}")
print(f"Solution of the system: x1 = {solution[0]}, x2 = {solution[1]}")

# Plotting the lines
x_vals = np.linspace(-10, 10, 400)
y1 = (1 - a_value * x_vals) / 2
y2 = (12 - 4 * x_vals) / 8

plt.plot(x_vals, y1, label=f'{a_value}x1 + 2x2 = 1')
plt.plot(x_vals, y2, label='4x1 + 8x2 = 12')
plt.plot(solution[0], solution[1], 'ro', label='Intersection Point')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Graph of the System of Equations')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

**Упражнение 2:** Найдите значение коэффициента $a$, при котором система не имеет решений (то есть чтобы функции были параллельны). Постройте линии, образованные вашими уравнениями, чтобы убедиться, что решения нет.

$$
\begin{eqnarray*}
 ax_1 + 2x_2 & = & 1 \\
4x_1 + 8x_2 & = & 12 \\
\end{eqnarray*}
$$

```python
# Define the value of a that makes the lines parallel
# For the lines to be parallel, the ratios of coefficients must be equal:
# a / 4 = 2 / 8 => a = 1
a = 1

# Define the equations:
# Equation 1: ax + 2y = 1 => x + 2y = 1
# Equation 2: 4x + 8y = 12

# Create x values
x = np.linspace(-10, 10, 400)

# Compute y values for both equations
y1 = (1 - a * x) / 2
y2 = (12 - 4 * x) / 8

# Plot the lines
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label=f'{a}x + 2y = 1')
plt.plot(x, y2, label='4x + 8y = 12', linestyle='--')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('System of Equations: No Solution (Parallel Lines)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

**Упражнение 3:** Вычислите $-2E$, $G+F$, $4F-G$, $HG$ и $GE$, используя следующие определения матриц. Сначала выполните упражнение на бумаге, а затем проверьте, выполнив расчет с массивами NumPy. Для умножения матриц используйте np.dot

$$
\begin{equation}
E = \left[ \begin{array}{r} 5 \\ -2 \end{array}\right] \hspace{1cm} 
F = \left[ \begin{array}{rr} 1 & 6 \\ 2 & 0 \\ -1 & -1 \end{array}\right] \hspace{1cm}
G = \left[ \begin{array}{rr} 2 & 0\\ -1 & 3 \\ -1 & 6 \end{array}\right] \hspace{1cm}
H = \left[ \begin{array}{rrr} 3 & 0 & 1 \\ 1 & -2 & 2 \\ 3 & 4 & -1\end{array}\right]
\end{equation}
$$

```python
# Define the matrices
E = np.array([[5], [-2]])
F = np.array([[1, 6], [2, 0], [-1, -1]])
G = np.array([[2, 0], [-1, 3], [-1, 6]])
H = np.array([[3, 0, 1], [1, -2, 2], [3, 4, -1]])

# Perform the operations
neg_2E = -2 * E
G_plus_F = G + F
fourF_minus_G = 4 * F - G
HG = np.dot(H, G)
GE = np.dot(G, E)

# Print the results
print("−2E =\n", neg_2E)
print("\nG + F =\n", G_plus_F)
print("\n4F − G =\n", fourF_minus_G)
print("\nHG =\n", HG)
print("\nGE =\n", GE)
```
