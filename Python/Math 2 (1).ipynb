```python
#1. Создайте функцию f(x) = x^6 + 10 * x^3 - 1000
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return x**6 + 10 * x**3 - 1000

# Generate x values over a suitable range
x = np.linspace(-10, 10, 400)

# Compute corresponding y values
y = f(x)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x^6 + 10x^3 - 1000')
plt.title('Plot of f(x) = x^6 + 10x^3 - 1000')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
#5. Найдите производную функции f(x) и создайте новую функцию на ее основе как f_diff(x)
```

```python
# Define the original function f(x)
def f(x):
    return x**6 + 10 * x**3 - 1000

# Define the derivative function f_diff(x)
def f_diff(x):
    return 6 * x**5 + 30 * x**2

# Generate x values
x_points = np.linspace(-10, 10, 400)

# Compute the derivative values
y_diff = f_diff(x_points)

# Plot the derivative
plt.figure(figsize=(10, 6))
plt.plot(x_points, y_diff, label="f'(x) = 6x⁵ + 30x²", color='green')
plt.title("Derivative of f(x) = x⁶ + 10x³ - 1000")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#8 Найдите уравнение касательной линии в точке x = 9
```

```python
# Define the function and its derivative
def f(x):
    return x**6 + 10*x**3 - 1000

def f_prime(x):
    return 6*x**5 + 30*x**2

# Point where we find the tangent
a = 9

# Calculate f(9) and f'(9)
f_a = f(a)
f_prime_a = f_prime(a)

# Tangent line equation
# y - f(a) = f'(a)(x - a)
# y = f'(a)(x - a) + f(a)
print(f"f(9) = {f_a}")
print(f"f'(9) = {f_prime_a}")

# Tangent line equation
print(f"Tangent line equation: y = {f_prime_a}(x - {a}) + {f_a}")
```

```python
#9 Покажите на графике функцию f(x) & касательную линию в точке x = 9
```

```python
# Define the function and its derivative
def f(x):
    return x**6 + 10*x**3 - 1000

def f_prime(x):
    return 6*x**5 + 30*x**2

# Point where we find the tangent
a = 9

# Calculate f(9) and f'(9)
f_a = f(a)
f_prime_a = f_prime(a)

# Define the tangent line function
def tangent_line(x):
    return f_prime_a * (x - a) + f_a

# Create x points for plotting
x_points = np.linspace(0, 12, 100)  # Range from 0 to 12 for better visualization
y_points = f(x_points)  # Function values
tangent_points = tangent_line(x_points)  # Tangent line values

# Plot the function and the tangent line
plt.plot(x_points, y_points, label='f(x) = x^6 + 10x^3 - 1000', color='blue')
plt.plot(x_points, tangent_points, label='Tangent line at x=9', color='red', linestyle='--')

# Mark the point of tangency
plt.scatter([a], [f_a], color='green')  # Point of tangency
plt.text(a, f_a, f'  (9, {f_a})', verticalalignment='bottom', horizontalalignment='right')

# Add titles and labels
plt.title('Function and Tangent Line at x = 9')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```
