#1. Вероятность того, что сегодня будет дождь составляет 30%, а вероятность того, что ваш заказ на зонт прибудет вовремя, составляет 40%. Вам не терпится прогуляться сегодня под дождем, и вы не можете без него!

Какова вероятность того, что пойдет дождь и ваш зонт прибудет вовремя?

```python
# Define the probabilities
p_rain = 0.3
p_umbrella_on_time = 0.4

# Calculate the joint probability of both independent events
p_both = p_rain * p_umbrella_on_time

# Output the result
print(f"The probability that it rains and the umbrella arrives on time is {p_both:.2f} or {p_both*100:.0f}%.")
```

#2. У вас есть закрытый ящик с 10 красными, 15 зелеными и 6 черными шарами. Если мы достанем 3 шара из коробки вслепую, какая вероятность, что первый будет красным, второй зеленым, третий черным?

```python
# Количество шаров
red = 10
green = 15
black = 6
total = red + green + black

# Вероятности по порядку без возвращения
p_red_first = red / total
p_green_second = green / (total - 1)
p_black_third = black / (total - 2)

# Общая вероятность
probability = p_red_first * p_green_second * p_black_third

print(f"Вероятность: {probability:.4f} или {probability * 100:.2f}%")
```

#3. Выполните эксперимент используя binom.rvs со 100 честными монетками, и 1.000 повторениями (size). Сохраните результат в results, и заметьте что если вы не указываете random_state то вы получите разные наборы данных.

```python
import numpy as np
from scipy.stats import binom

# Perform the binomial experiment
# 100 fair coin flips (n=100), probability of heads (p=0.5), repeated 1000 times
results = binom.rvs(n=100, p=0.5, size=1000)

# Display the first few results
print("First 10 results of the experiment:", results[:10])
```

#4. Покажите results на гистограмме

```python
import matplotlib.pyplot as plt

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(results, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of 1000 Experiments with 100 Fair Coin Flips')
plt.xlabel('Number of Heads')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
