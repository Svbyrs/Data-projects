```python
# Define the function f(x)
def f(x):
    return 3 * x**2 + 2 * x - 10
```

```python
#2. Создайте список x_points (или массив через numpy), в котором у нас лежат числа от -10 до 10
```

```python
import numpy as np

# Create an array of numbers from -10 to 10
x_points = np.arange(-10, 11)

print(x_points)
```

```python
#3. Используйте x_points как аргумент для созданной функции f(x), и выведите на экран результаты функции для каждого значения 
# x_points, также сохраните результаты в y_points (cписок или массив)
```

```python
# Define the function
def f(x):
    return 3 * x**2 + 2 * x - 10

# Create x_points from -10 to 10
x_points = np.arange(-10, 11)

# Apply the function to each element in x_points
y_points = f(x_points)

# Print the results
print("x_points:", x_points)
print("y_points:", y_points)
```

```python
#4. Покажите на графике x_points & y_points
```

```python
import matplotlib.pyplot as plt

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(x_points, y_points, marker='o', linestyle='-', color='blue', label='f(x) = 3x² + 2x - 10')
plt.title('Graph of f(x) = 3x² + 2x - 10')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#5. Найдите производную функции f(x) и создайте новую функцию на ее основе как f_diff(x)
```

```python
import sympy as sp

# Define the symbolic variable
x = sp.symbols('x')

# Define the original function f(x)
f_expr = 3 * x**2 + 2 * x - 10

# Compute the derivative of f(x)
f_derivative = sp.diff(f_expr, x)

# Define the new function f_diff(x) based on the derivative
def f_diff(x_val):
    return 6 * x_val + 2

# Display the symbolic derivative
f_derivative
```

```python
# Compute y values for both functions
y_points = f(x_points)
y_diff_points = f_diff(x_points)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_points, y_points, label='f(x) = 3x² + 2x - 10', marker='o')
plt.plot(x_points, y_diff_points, label="f'(x) = 6x + 2", linestyle='--', marker='x')
plt.title('График функции f(x) и её производной f\'(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
#8. Найдите уравнение касательной линии в точке x = 8
```

```python
# Point of tangency
x0 = 8
y0 = f(x0)
slope = f_diff(x0)

# Equation of the tangent line: y = m*x + b
# We solve for b using y0 = m*x0 + b => b = y0 - m*x0
intercept = y0 - slope * x0

# Construct the equation of the tangent line
tangent_equation = f"y = {slope}x + {intercept}"

tangent_equation
```

```python
#9. Покажите на графике функцию f(x) & касательную линию в точке x = 8
```

```python
# Define the tangent line at x = 8
def tangent_line(x):
    return 50 * x - 202

# Generate x values
x_points = np.linspace(-10, 10, 400)
y_points = f(x_points)
tangent_y = tangent_line(x_points)

# Plot the function and the tangent line
plt.figure(figsize=(10, 6))
plt.plot(x_points, y_points, label='f(x) = 3x² + 2x - 10', color='blue')
plt.plot(x_points, tangent_y, label='Tangent at x = 8: y = 50x - 202', color='orange', linestyle='--')

# Highlight the point of tangency
x_tangent = 8
y_tangent = f(x_tangent)
plt.scatter(x_tangent, y_tangent, color='red', zorder=5)
plt.text(x_tangent, y_tangent + 10, f'({x_tangent}, {y_tangent})', ha='center')

# Add labels and legend
plt.title('Function f(x) and Tangent Line at x = 8')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#10. Найдите мин или макс функции f(x) и покажите на графике касательную линию, которая проходит через этот мин или макс функции
```

```python
# Find the vertex (minimum point) of the parabola
# For f(x) = ax^2 + bx + c, the vertex x = -b / (2a)
a, b = 3, 2
x_vertex = -b / (2 * a)
y_vertex = f(x_vertex)

# Tangent line at the vertex
slope = f_diff(x_vertex)
intercept = y_vertex - slope * x_vertex

# Define the tangent line function
def tangent_line(x):
    return slope * x + intercept

# Generate x values for plotting
x_vals = np.linspace(-10, 10, 400)
y_vals = f(x_vals)
tangent_vals = tangent_line(x_vals)

# Plot the function and the tangent line
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = 3x² + 2x - 10', color='blue')
plt.plot(x_vals, tangent_vals, label='Tangent at minimum', linestyle='--', color='orange')
plt.scatter(x_vertex, y_vertex, color='red', zorder=5, label=f'Minimum point ({x_vertex:.2f}, {y_vertex:.2f})')
plt.title('Function and Tangent at Minimum Point')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#11. Посчитайте интеграл функции f(x) между 0 и 5. Также отобразите на графике площадь, которую мы посчитали 
```

```python
from scipy.integrate import quad

# Calculate the definite integral from 0 to 5
integral_value, _ = quad(f, 0, 5)

# Generate x values for plotting
x_vals = np.linspace(-2, 6, 400)
y_vals = f(x_vals)

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label='f(x) = 3x² + 2x - 10', color='blue')

# Fill the area under the curve between x=0 and x=5
x_fill = np.linspace(0, 5, 200)
y_fill = f(x_fill)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5, label='Area under f(x) from 0 to 5')

# Annotate the integral value
plt.title(f"Definite Integral of f(x) from 0 to 5 = {integral_value:.2f}")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
